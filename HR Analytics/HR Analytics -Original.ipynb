```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, matthews_corrcoef, \
    roc_auc_score,precision_score, recall_score, \
    average_precision_score

def get_scoring_functions():
    cfuncs = [accuracy_score, f1_score, matthews_corrcoef, recall_score]
    pfuncs = [roc_auc_score, average_precision_score]
    return cfuncs, pfuncs

def get_tuple_cols():
    cfuncs, pfuncs = get_scoring_functions()
    return ['model'] + [f.__name__ for f in cfuncs] + [f.__name__ for f in pfuncs]

def get_scores(model_name, y_true, y_preds):
    cfuncs, pfuncs = get_scoring_functions()

    cscores = {f.__name__: f(y_true, y_preds) for f in cfuncs}
    pscores = {f.__name__: f(y_true, y_preds) for f in pfuncs}

    d = {**cscores, **pscores}
    d['model'] = model_name

    return tuple([d[c] for c in get_tuple_cols()])

models = [Log,RF, DT, NB, SVM, Xg, SGD, AB]
model_names = [type(m).__name__ for m in models]
print("\033[5m"+ "BEFORE APPLYING HYPER PARAMETERS"+"\033[0m\n")
y_preds = {type(model).__name__: model.predict(X_test) for model in models}
scores = [get_scores(name, Y_test, y_preds[name]) for name in model_names]
df = pd.DataFrame(scores, columns=get_tuple_cols())
df=df.sort_values('accuracy_score',ascending=False).reset_index(drop=True)
df=df.style.background_gradient(cmap='viridis')
df
```

## HYPER PARAMETER TUNING RANDOM FOREST 

```python
model_comparison = {}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_estimators': [100, 200, 300, 500],
    'criterion': ['gini','entropy'],
    'max_depth': [None, 10, 20, 30, 40,50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]}    
    
```

```python
model_grid = GridSearchCV(RF, param_grid=params, cv=5)
model_grid.fit(X_train_resampled, Y_train_resampled)
```

```python
 best_rf =model_grid.best_params_
```

```python
print(best_rf)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_grid_RF = RandomForestClassifier(criterion= 'gini', max_depth= 30, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 200, random_state=42)
```

```python
model_gridRF=model_grid_RF.fit(X_train_resampled, Y_train_resampled)
Y_pred_gridRF = model_grid_RF.predict(X_test)
```

```python
cv_accuracy_RF = cross_val_score(estimator = model_gridRF, X = X_train_resampled, y = Y_train_resampled, cv = 5) 
print(cv_accuracy_RF.mean())
```

```python
print(accuracy_score(Y_test, Y_pred_gridRF))
model_comparison['Random Forest']=[accuracy_score(Y_pred_gridRF,Y_test),f1_score(Y_pred_gridRF,Y_test,average='weighted'), cv_accuracy_RF.mean()]
```

```python
print(classification_report(Y_test, Y_pred_gridRF))
```

```python
print(confusion_matrix(Y_test, Y_pred_gridRF))
```
