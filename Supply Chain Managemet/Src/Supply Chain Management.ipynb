```python
# splitting in dependent and independent variable
```

```python
#independent variables
features=["national_inv","lead_time","sales_3_month","sales_6_month","forecast_6_month","forecast_9_month"]
```

```python
X=df2[features]
Y=df2["went_on_backorder"]
```

```python
# apply minmax scaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
X=pd.DataFrame(X,columns=features)
```

```python
X.head()
```

```python
X1=X.copy()
Y1=Y.copy()
```

```python
# splitting the data in training and testing
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=110)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# model building
```

```python
from sklearn.metrics import roc_auc_score,confusion_matrix,accuracy_score,classification_report
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
Y_pred_dt=dt.predict(X_test)
```

```python
# model performance
```

```python
print(dt.score(X_train,Y_train))
print("***********\n")
print(accuracy_score(Y_test,Y_pred_dt))
print("***********\n")
print(confusion_matrix(Y_test,Y_pred_dt))
print("***********\n")
print(classification_report(Y_test,Y_pred_dt))
```

```python
# randomforest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,Y_train)
Y_pred_rf=rf.predict(X_test)
print(rf.score(X_train,Y_train))
print("***********\n")
print(accuracy_score(Y_test,Y_pred_rf))
print("***********\n")
print(confusion_matrix(Y_test,Y_pred_rf))
print("***********\n")
print(classification_report(Y_test,Y_pred_rf))
```
